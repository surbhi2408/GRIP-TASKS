```python
# Importing required libraries
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# loading the dataset
df = pd.read_csv("globalterrorism.csv", encoding="ISO-8859-1")
df
```

```python
# depicts the datatypes and the null values
df.info()
```

```python
# checking for null values
df.isnull().sum()
```

```python
# checking for duplicates
df.duplicated().sum()
```

```python
df.dropna(axis=1, how="all")
```

```python
# rename the columns which aren't well labelled
df.rename(columns={"iyear" : "Year", "imonth" : "Month", "iday" : "Day", "country_txt" : "Country",
                   "region_txt" : "Region", "provstate" : "State", "latitude" : "Latitude", 
                   "longitude" : "Longitude", "summary" : "Summary", "attacktype1_txt" : "AttackType", 
                   "targtype1_txt" : "TargetType", "gname" : "Group", "weaptype1_txt" : "WeaponType", 
                   "nkill" : "Kill", "nwound" : "Wound",}, inplace=True)
```

```python
df.head()
```

```python
# Keeping only the columns which can be responsible for terrorist attacks and which should be taken into account
df = df[["Year", "Month", "Day", "AttackType", "TargetType", "Country", "Region",
         "Latitude", "Longitude", "Summary", "State", "motive", "WeaponType", 
         "Kill", "Wound", "Group", "Summary"]]
df.head()
```

```python
# Loading the correlation
corr = df.corr()
corr
```

```python
# Plot heatmap for collinearity
plt.figure(figsize=(20,10))
sns.heatmap(corr,annot=True)
```

## Exploratory Data Analysis

```python
print("Year with the most terroist attacks :", df["Year"].value_counts().idxmax())
print("Month with the most terroist attacks :", df["Month"].value_counts().idxmax())
print("Most terroist attakcs types :", df["AttackType"].value_counts().idxmax())
```

```python
print("Country with the highest terroist attacks :", df["Country"].value_counts().idxmax())
```

```python
print("Country with the highest terroist attacks :", df["Country"].value_counts().idxmax())
print("Region with the highest terroist attacks :", df["Region"].value_counts().idxmax())
print("Group with the most terroist attacks :", df["Group"].value_counts().index[0])
```

```python
# Since unknown is something this is not known, we should try this
print("Group with the most terroist attacks :",df["Group"].value_counts().index[1])
```

## Various Attacking methods

```python
plt.figure(figsize=(20,10))
sns.countplot(x="AttackType",data=df,order=df["AttackType"].value_counts().index)
plt.xticks(rotation=90)
plt.title("Various attacking methods")
plt.show()
```

## What are the major terroist targets?

```python
plt.figure(figsize=(20,10))
sns.countplot(x="TargetType", data=df, order=df["TargetType"].value_counts().index)
plt.xticks(rotation=90)
plt.title("Major Terroist Targets")
plt.show()
```

## Which regions are the most affected terroist targets?

```python
plt.figure(figsize=(20,10))
sns.countplot(x="Region", data=df, order=df["Region"].value_counts().index)
plt.xticks(rotation=90)
plt.title("Most affected terroist targets region wise")
plt.show()
```

## Which countries are the most affected countries by terrorism?

```python
plt.figure(figsize=(20,10))
x=df["Country"].value_counts().index
y=df["Country"].value_counts().values
sns.barplot(x,y)
plt.title("Countries affected by terrorism")
plt.show()
```

```python
# since the countries are not visible, let's do it till the top 15 countries
plt.figure(figsize=(20,10))
x=df["Country"].value_counts()[:15].index
y=df["Country"].value_counts()[:15].values
sns.barplot(x,y)
plt.title("Countries affected by Terrorism")
plt.show()
```

## No. of people got killed country wise

```python
plt.figure(figsize=(20,10))
x=df["Country"].value_counts()[:15].index
y=df["Kill"].value_counts()[:15].values
sns.barplot(x,y)
plt.title("Countries with no. of people killed in terroist attacks")
plt.show()
```

## No. of terrorist attacks with respect to year

```python
plt.figure(figsize=(20,10))
x=df["Year"].value_counts().index
y=df["Year"].value_counts().values
sns.barplot(x,y)
plt.show()
```
